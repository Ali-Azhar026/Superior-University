```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("train.csv")
df
```

```python
print(f"Number of Rows: {df.shape[0]}\nNumber of Columns: {df.shape[1]}")
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
def fillNaMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Age', 'Cabin', 'Embarked']
fillNaMode(columns)
```

```python
df.isnull().sum()
```

```python
def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Age', 'Fare']
changeFloattoInt64(columns)
```

```python
df.info()
```

```python
def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))
        df[i] = data_LE.transform(df[i])

columns = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']
encodeCols(columns)
```

```python
df.info()
```

# Training

```python
X = df.drop(columns=['Survived'])
y = df['Survived']                 
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=50, max_depth=3, learning_rate=0.1, random_state=42)
model.fit(X, y)
```
