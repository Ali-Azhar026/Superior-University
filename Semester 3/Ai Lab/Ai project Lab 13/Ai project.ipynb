```python
def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['tests_per_case', 'reproduction_rate', 'stringency_index', 'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 'gdp_per_capita', 'extreme_poverty', 'cardiovasc_death_rate', 'diabetes_prevalence' ]
changeFloattoInt64(columns)
```

```python
df.info()
```

```python
def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))
        df[i] = data_LE.transform(df[i])

columns = ['date', 'continent', 'location', 'positive_rate']
encodeCols(columns)
```

# Data Splitting / Train-Test Split

```python
X = df.drop(columns=['new_deaths'])
y = df['new_deaths']                 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
print("TrainX:",X_train.shape)
print("TrainY:",y_train.shape)
print("TestX:",X_test.shape)
print("TestY:",y_test.shape)
```

# Training/Apply Regressor

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
import pickle

pickle.dump(model, open("model.pkl", "wb"))
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print('Accuracy Score of Training Data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
