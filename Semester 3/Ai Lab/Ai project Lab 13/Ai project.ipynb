# Handel Missing values

```python
df = df.dropna(subset=['new_deaths'])
```

```python
df['new_deaths'] = df['new_deaths'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.info()
```

```python
def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['tests_per_case', 'reproduction_rate', 'stringency_index', 'population_density', 'median_age', 'aged_65_older', 'aged_70_older', 'gdp_per_capita', 'extreme_poverty', 'cardiovasc_death_rate', 'diabetes_prevalence' ]
changeFloattoInt64(columns)
```

```python
df.info()
```

```python
categorical_columns = ['date', 'continent', 'location', 'positive_rate']
le = LabelEncoder()
for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))
```

# Data Splitting / Train-Test Split

```python
X = df.drop(columns=['new_deaths'])
y = df['new_deaths']                 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)  
```

```python
print("TrainX:",X_train.shape)
print("TrainY:",y_train.shape)
print("TestX:",X_test.shape)
print("TestY:",y_test.shape)
```

# Gradient Boosting Classifer

```python
model = GradientBoostingClassifier(n_estimators=50, max_depth=3, learning_rate=0.1, random_state=42)
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
model.fit(X_train, y_train)
```

```python
y_train_prediction = model.predict(X_train)
y_test_prediction = model.predict(X_test)
```

```python
train_accuracy = accuracy_score(y_train, y_train_prediction)
test_accuracy = accuracy_score(y_test, y_test_prediction)
```

```python
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```
